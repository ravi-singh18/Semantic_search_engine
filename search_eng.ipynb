```python
import chromadb
client = chromadb.PersistentClient(path=r"C:\Users\errav\genai\search_engine\chromadb4")
collection = client.get_or_create_collection(name="search_engine_project_3", metadata={"hnsw:space": "cosine"})
```

```python
df.reset_index()
```

```python
for i, embedding in enumerate(chunk_emb):
    collection.add(ids=f"chunk_{i}",  # Replace with a more descriptive ID scheme if needed
                   embeddings=embedding.tolist(),  # Convert to a list for Chroma
                   documents=df.loc[i, "file_content_chunks"],  # Use chunk_text as documents
                   metadatas={"name": df.loc[i, "name"]}  # Use 'name' column data as metadata
    )
```

```python
import re
def clean(text):
    text = text.lower()
    text = re.sub("[^\w\d]"," ",text)
    return text
```

```python
subtitle = 'Princess of spanish cinema princess yeah here dear come on fontiveros says you should eat there s a lot of work ahead sure can t slow the shoot down can we and these people these stay out of politics look what happened to your father i m not hungry that bitch took my appetite away so here this is for you watch it macarena what s the matter what s i know you as if you were mine you like the jew he s russian well whatever but if he s a prisoner he must ve done something hey back don t push me or i '
subtitle_cleaned =clean(subtitle)
```

```python
query_embedding = model.encode([subtitle_cleaned])
```

```python
# Perform a query on the ChromaDB collection using the embeddings
searched_result = collection.query(
    query_embeddings=query_embedding,
    n_results=10,
    include=['documents']
)
```

```python
documents = searched_result['documents']

# Iterate over the documents and print each document
for i, query_documents in enumerate(documents):
    for j, document in enumerate(query_documents):
        print(f"Document {i+1}, Item {j+1}: {document}")

```
